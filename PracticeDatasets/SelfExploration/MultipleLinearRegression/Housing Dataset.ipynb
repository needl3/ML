
## Multiple Linear Regression

### Splitting to training and testing data

```python
import random
train = data.sample(frac=0.8, random_state=20)
test = data.drop(index=train.index)
print(len(train.index), len(test.index))
```

```python
xTrain = train.drop(columns = "Price")
yTrain = train["Price"]

xTest = test.drop(columns = "Price")
yTest = test["Price"]
xTrain.head()
```

## Feature scaling for input variables

```python
def scale(data):
    for col in list(data.columns):
        data[col] = (data[col]-data[col].mean())/(data[col].max() - data[col].min())
scale(xTrain)
scale(xTest)
xTrain.head()
```

```python
def dot(wVec,xVec):
    s = 0
    for i in range(len(wVec)):
        s += wVec[i]*xVec[i]
    return s
```

```python
# We can wrap everything from this section into fit() function that will
# Apply gradient descent and find optimum wVec and bias
wVec = [0 for _ in range(len(xTrain.columns))]
bias = 0
lrate = 0.5
predicted = [dot(wVec, xTrain.iloc[i]) + bias for i in range(len(xTrain.index))]

errors = []
for i in range(len(predicted)):
    diff = predicted[i]-yTrain.iloc[i]
    errors.append(diff)

# Directly changed because it's not needed anymore in further operations
# I already stored all possible predictions and errors
bias = bias - lrate * 1/len(yTrain.index) * sum(errors)

for i in range(len(wVec)):
    iSum = sum([errors[j] * xTrain.iloc[j,i] for j in range(len(xTrain.index))])
    wVec[i] = wVec[i] - lrate * 1/len(xTrain.index) * iSum

print(wVec, bias)
```

### Now make a fit function that does needed amount of iteration for gradient descent to find optimal solution 

```python
def fit(xTrain, yTrain, plot=False):
    wVec = [0 for _ in range(len(xTrain.columns))]
    bias = 0
    lrate = 0.05
    total_error_history = []
    n = 0
    while(len(total_error_history) < 2 or (total_error_history[-2]-total_error_history[-1] > 0.000002)):
        predicted = [dot(wVec, xTrain.iloc[i, :]) + bias for i in range(len(xTrain.index))]

        errors = []
        for i in range(len(predicted)):
            diff = predicted[i]-yTrain.iloc[i]
            errors.append(diff)

        # Directly changed because it's not needed anymore in further operations
        # I already stored all possible predictions and errors
        bias = bias - lrate * 1/len(yTrain.index) * sum(errors)

        for i in range(len(wVec)):
            iSum = sum([errors[j] * xTrain.iloc[j,i] for j in range(len(xTrain.index))])
            wVec[i] = wVec[i] - lrate * 1/len(xTrain.index) * iSum

        total_error_history.append(abs(sum(errors)))
        print(n, " Iterations complete with total error: ", abs(sum(errors)))
        n+=1

    if plot:
        plt.plot(total_error_history)
        plt.show()
    return (wVec, bias)

vecs,bias = fit(xTrain, yTrain, plot=True)
```

```python
predictedValues = [dot(vecs, xTest.iloc[i]) + bias for i in range(len(xTest.index))]
plt.scatter(xTest.iloc[:, 0], yTest)
plt.scatter(xTest.iloc[:, 0], predictedValues)
plt.show()
```
