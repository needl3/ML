```python
import pandas as pd
import numpy as np
```

```python
sales1 = pd.read_csv('https://storage.googleapis.com/codehub-data/1-A-2-4-sales1.csv')
sales2 = pd.read_csv('https://storage.googleapis.com/codehub-data/1-A-2-4-sales2.csv')
```

```python
sales1.info()
sales2.info()
```

```python
sales1.head()
sales2.head()
```

```python
sales = pd.merge(sales1, sales2, on="Order ID")
sales.head()
```

```python
sales['Country'] = sales.Country.fillna('unknown')
sales["Order Priority"] = sales["Order Priority"].fillna(sales["Order Priority"].mode()[0])
sales["Order Priority"].mode()
```

```python
sales = sales.fillna(sales.mean())
```

```python
sales.info()
```

```python
sales.Region = sales.Region.fillna('unknown')
sales.Region.info()
```

```python
sales.info()
```

```python
summary = sales.groupby(by="Item Type", as_index=True)['Unit Cost'].agg(['count','mean','std'])
summary
```

```python
summary.loc[['Fruits', 'Vegetables', 'Baby Food', 'Meat']]
```

```python
sales["Total Profit"] = sales["Total Revenue"] - sales["Total Cost"]
sales.head()
```

```python
(sales[sales["Total Profit"]/sales["Total Cost"]*100 >= 60])
```

```python
sales["Total Profit%"]=sales["Total Profit"]/sales["Total Cost"]*100
sales[sales["Total Profit%"] >= 60].count().iloc[0]
```

```python
sales.groupby(by="Item Type", as_index=True)["Unit Cost"].agg(["mean", "count","std"])
```

```python
def calculateProfit(unitCost, unitPrice, unitSold):
    return unitSold * (unitPrice-unitCost)
```

```python
pd.concat([sales["Total Profit"], sales.apply(lambda x: calculateProfit(x["Unit Cost"], x["Unit Price"], x["Units Sold"]), axis=1)], axis=1)
```
